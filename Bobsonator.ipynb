```python
from PIL import Image, ImageFont, ImageDraw 
import random
pic = Image.new('RGBA', (570,380), (7,8,152))
resigners = ['Boris Johnson', 'Michael Gove', 'Rory Stewart', 'James Cleverly', 'Andrea Leadsom', 'Matt Hancock', 'Dominic Raab', 'Kit Malthouse', 'Esther McVey', 'Sajid Javid', 'Jeremy Hunt', 'Mark Harper', 'James Cleverly']
font = ImageFont.truetype("/home/xtophercook/fonts/PressStart2P.ttf", 14)
namelist = ["Onson Sweemey", "Darryl Archideld", "Anatoli Smorin", "Rey McSriff", "Glenallen Mixon", "Mario McRlwain", "Raul Chamgerlain", "Kevin Nogilny", "Tony Smehrik", "Willie Dustice", "Jeromy Gride", "Scott Dourque", "Shown Furcotte", "Dean Wesrey", "Mike Truk", "Tim Sandaele", "Karl Dandleton", "Mike Sernandez", "Todd Bonzalez"]
essentials =  ["Sleve McDichael", "Dwigt Rortugal", "Bobson Dugnutt"]

if (19-len(resigners)) < 0:
    number = 0
else:
    number = 19-len(resigners)
names = random.sample(namelist, number) + resigners + essentials

if (len(names) % 2 != 0):
    names = random.sample(namelist, number+1) + resigners + essentials
else:
    random.sample(namelist, number) + resigners + essentials
    
random.shuffle(names)

start = 30
indent = 30
indent_2 = (pic.size[0]/2 + indent)

halfway = int(len(names)/2)

i = start
for entry in names[:halfway]:
    ImageDraw.Draw(pic).text((indent + 3, i + 1), entry.upper(),(0,0,0), font=font)
    ImageDraw.Draw(pic).text((indent, i), entry.upper(),(255,255,255), font=font)
    i = i+20
    print(i)

i = start
for entry in names[halfway:]:
    ImageDraw.Draw(pic).text((indent_2 + 3, i + 1), entry.upper(),(0,0,0), font=font)
    ImageDraw.Draw(pic).text((indent_2, i), entry.upper(),(255,255,255), font=font)
    i = i+20

length = i+start-10
area = (0, 0, 570, length)
cropped_img = pic.crop(area)

cropped_img
```
